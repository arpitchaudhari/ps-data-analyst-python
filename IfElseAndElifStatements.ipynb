# Iterating over Dictionary Elements using a for loop

```python
names_and_ages = {
    'John':53,
    'Alice':57,
    'Bill':14,
    'Dave':25
}
```

```python
for key in names_and_ages:
    print('Name: ',key, 'Age: ',names_and_ages[key])
```

**Item function** returns key-value pairs in dictionary in the form of list of tuples

```python
names_and_ages.items()
```

```python
for key,val in names_and_ages.items():#Iterating over list of tuples
    print(key,val)
```

# Conditional Looping Using While Loops

```python
a = 5
while a < 10:
    print(a)
```

```python
b = 7
while b == 7:
    print(b)
    b=b*2
```

# Break, Continue and pass keywords to control execution
* continue - skip this iteration of loop
* pass - No code here just carry on (Want to do nothing)

```python
data = [1,2,3,4,5]
```

```python
for d in data:
    print(d)
    
    if d == 3:
        print('Found')
        break
```

```python
 for d in data:
    if d == 3:        
        continue

    print(d)
```

```python
for i in range(10,21):
    if i >= 13 and i <= 16:
        pass
    else:
        print(i)
```
